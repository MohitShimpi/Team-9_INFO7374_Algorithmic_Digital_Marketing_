```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import datetime as dt
import numpy as np
```

```python
data = pd.read_csv('User_Data.csv')
```

```python
data
```

```python
data["Subscription_StartDate"] = pd.to_datetime(data['Subscription_StartDate'], utc=False)
```

```python
data
```

```python
uk_data_group = data.groupby('user_id').agg({'Subscription_StartDate': lambda date: (date.max() - date.min()).days,
                                        'Subscription_EndDate': lambda num: len(num),
                                        'Subscription_Cost': lambda price: price.sum()})
```

```python
uk_data_group
```

```python
# Change the name of columns
uk_data_group.columns=['num_days','num_transactions','spent_money']
uk_data_group.head()
```

```python
uk_data_group['avg_order_value'] = uk_data_group['spent_money']/uk_data_group['num_transactions']
```

```python
uk_data_group.head()
```

```python
purchase_frequency = sum(uk_data_group['num_transactions'])/uk_data_group.shape[0]
```

```python
# Repeat Rate
repeat_rate=uk_data_group[uk_data_group.num_transactions > 74].shape[0]/uk_data_group.shape[0]
```

```python
uk_data_group[uk_data_group.num_transactions > 75].shape[0]
```

```python
repeat_rate
```

```python
#Churn Rate
churn_rate=1-repeat_rate
```

```python
churn_rate
```

```python
purchase_frequency,repeat_rate,churn_rate
```

```python
uk_data_group['profit_margin']=uk_data_group['spent_money']*0.05
```

```python
# Customer Value
uk_data_group['CLV']=(uk_data_group['avg_order_value']*purchase_frequency)/churn_rate
```

```python
#Customer Lifetime Value
uk_data_group['cust_lifetime_value']=uk_data_group['CLV']*uk_data_group['profit_margin']
```
